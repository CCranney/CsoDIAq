# Determing Optimal Match Number for FDR Calculations (library peak max = 31)

```python
import pandas as pd
import re
from matplotlib import pyplot
pd.set_option('display.max_rows', None)
```

## Loading Results

```python
def make_graph_df(listOfFiles):
    concatLists = []
    for file in listOfFiles:
        temp = pd.read_csv(file)
        groupName = re.sub('csodiaq_lib-iproph1-(\d+)peaks_exp-n1b(_sqrt)?(_corrected1)_FDRGraph_(\w+).csv', r'lib\1\2\3', file)
        temp['group'] = [groupName for x in range(len(temp))]
        concatLists.append(temp)
    return pd.concat(concatLists)
```

```python
spec1 = 'csodiaq_lib-iproph1-20peaks_exp-n1b_corrected1_FDRGraph_spectral.csv'
spec2 = 'csodiaq_lib-iproph1-20peaks_exp-n1b_sqrt_corrected1_FDRGraph_spectral.csv'
spec3 = 'csodiaq_lib-iproph1-31peaks_exp-n1b_corrected1_FDRGraph_spectral.csv'
spec4 = 'csodiaq_lib-iproph1-31peaks_exp-n1b_sqrt_corrected1_FDRGraph_spectral.csv'
spectrum = make_graph_df([spec1, spec2, spec3, spec4])
spectrum
```

```python
pep1 = 'csodiaq_lib-iproph1-20peaks_exp-n1b_corrected1_FDRGraph_peptide.csv'
pep2 = 'csodiaq_lib-iproph1-20peaks_exp-n1b_sqrt_corrected1_FDRGraph_peptide.csv'
pep3 = 'csodiaq_lib-iproph1-31peaks_exp-n1b_corrected1_FDRGraph_peptide.csv'
pep4 = 'csodiaq_lib-iproph1-31peaks_exp-n1b_sqrt_corrected1_FDRGraph_peptide.csv'
peptide = make_graph_df([pep1, pep2, pep3, pep4])
peptide
```

```python
pro1 = 'csodiaq_lib-iproph1-20peaks_exp-n1b_corrected1_FDRGraph_protein.csv'
pro2 = 'csodiaq_lib-iproph1-20peaks_exp-n1b_sqrt_corrected1_FDRGraph_protein.csv'
pro3 = 'csodiaq_lib-iproph1-31peaks_exp-n1b_corrected1_FDRGraph_protein.csv'
pro4 = 'csodiaq_lib-iproph1-31peaks_exp-n1b_sqrt_corrected1_FDRGraph_protein.csv'
protein = make_graph_df([pro1, pro2, pro3, pro4])
protein
```

## Optimal FDR Cutoff

### Spectral Level Analysis

```python
spectrumFDR = spectrum.pivot(index='matches', columns='group', values='FDRCutoff')
spectrumFDR.plot()
```

### Peptide Level Analysis

```python
peptideFDR = peptide.pivot(index='matches', columns='group', values='FDRCutoff')
peptideFDR.plot()
```

### Protein Level Analysis

```python
proteinFDR = protein.pivot(index='matches', columns='group', values='FDRCutoff')
proteinFDR.plot()
```

## Percentage FDR Representation

### Data Rearrangement

```python
spectrum = spectrum.reset_index(drop=True)
peptide = spectrum.reset_index(drop=True)
protein = protein.reset_index(drop=True)

spectrum['%'] = [(spectrum['FDRCutoff'].loc[i]/spectrum['total'].loc[i]) for i in range(len(spectrum))]
peptide['%'] = [(peptide['FDRCutoff'].loc[i]/peptide['total'].loc[i]) for i in range(len(peptide))]
protein['%'] = [(protein['FDRCutoff'].loc[i]/protein['total'].loc[i]) for i in range(len(protein))]
```

### Spectrum Level Analysis

```python
spectrumPerc = spectrum.pivot(index='matches', columns='group', values='%')
spectrumPerc.plot()
```

### Peptide Level Analysis

```python
peptidePerc = peptide.pivot(index='matches', columns='group', values='%')
peptidePerc.plot()
```

### Protein Level Analysis

```python
proteinPerc = protein.pivot(index='matches', columns='group', values='%')
proteinPerc.plot()
```
